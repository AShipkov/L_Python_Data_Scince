# Функция обработки данных

```python
def prepare(df, dummy_features=None):
    df.fillna(0)
    
    idx = (df['HouseYear'] >= 2019)
    df.loc[idx, 'HouseYear'] = 2019
    
    Square = dict(df.groupby('Rooms')['Square'].median())
    idx = (df['Square'] < 10) & (df['Rooms'] < 2)
    df.loc[idx, 'Square'] = df.loc[idx, 'Rooms'].apply(lambda x: Square[x])
    
    idx = (df['Square'] < 30) & (df['Rooms'] > 1)
    df.loc[idx, 'Square'] = df.loc[idx, 'Rooms'].apply(lambda x: Square[x])
    
    idx = (df['Square'] < df['LifeSquare'])
    df.loc[idx, 'Square'] = df.loc[idx, 'LifeSquare']

    LifeSquare = dict(df.groupby('Rooms')['LifeSquare'].median())
    idx = (df['LifeSquare'].isnull())
    df.loc[idx, 'LifeSquare'] = df.loc[idx, 'Rooms'].apply(lambda x: LifeSquare[x])

    idx = (df['Rooms'] < 1)
    df.loc[idx, 'KitchenSquare'] = 0.0
    
    KitchenSquare = dict(df.groupby('Rooms')['KitchenSquare'].median())
    idx = (df['Rooms'] > 0)
    df.loc[idx, 'KitchenSquare'] = df.loc[idx, 'Rooms'].apply(lambda x: KitchenSquare[x])
    
    idx = (df['HouseFloor'] < df['Floor'])
    df.loc[idx, 'HouseFloor'] = df.loc[idx, 'Floor']
    
    df['Healthcare_1'].fillna(df['Healthcare_1'].median(), inplace=True)
    
    df['Ecology_2'] = df['Ecology_2'].apply(lambda x: 1 if x == 'A' else 0)
    df['Ecology_3'] = df['Ecology_3'].apply(lambda x: 1 if x == 'A' else 0)
    df['Shops_2'] = df['Shops_2'].apply(lambda x: 1 if x == 'A' else 0)
    
    if dummy_features != None:
        df = pd.get_dummies(data=df, columns=dummy_features, drop_first=True)
    
    return df
```

```python
train_data = prepare(clear_data, dummy_features=None)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
```

```python
X = train_data.drop(labels=['Price'], axis=1)
X = pd.get_dummies(data=X, columns=['DistrictId', 'Ecology_2', 'Ecology_3', 'Shops_2'], drop_first=True)
y = train_data['Price'].values
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_train = lr.predict(X_train_scaled)
y_pred_valid = lr.predict(X_valid_scaled)
```

```python
r2_score(y_train, y_pred_train), r2_score(y_valid, y_pred_valid)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred_train = rf.predict(X_train_scaled)
y_pred_valid = rf.predict(X_valid_scaled)
```

```python
r2_score(y_train, y_pred_train), r2_score(y_valid, y_pred_valid)
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train_scaled, y_train)
y_pred_train = knn.predict(X_train_scaled)
y_pred_valid = knn.predict(X_valid_scaled)
```

```python
r2_score(y_train, y_pred_train), r2_score(y_valid, y_pred_valid)
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train_scaled, y_train)
y_pred_train = dt.predict(X_train_scaled)
y_pred_valid = dt.predict(X_valid_scaled)
```

```python
r2_score(y_train, y_pred_train), r2_score(y_valid, y_pred_valid)
```

Выбираем RandomForestRegressor

```python
X = train_data.drop(labels=['DistrictId', 'Price'], axis=1)
y = train_data['Price'].values
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
```

```python
rf = RandomForestRegressor(random_state=42, n_jobs=-1)
random_grid = {
    'n_estimators': np.arange(200, 501, 20),
    'max_depth': np.arange(2, 51, 2),
    'max_features': [0.5, 0.6, 0.7, 0.8, 0.9],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10]
}
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=random_grid,
    n_iter=50,
    scoring='r2',
    cv=10,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
random_search.fit(X_train_scaled, y_train)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

Лучшие параметры: {'n_estimators': 360,  'min_samples_split': 5, 'min_samples_leaf': 2,'max_features': 0.6, 'max_depth': 32}

```python
y_pred_train = random_search.best_estimator_.predict(X_train_scaled)
y_pred_valid = random_search.best_estimator_.predict(X_valid_scaled)
```

```python
r2_score(y_train, y_pred_train), r2_score(y_valid, y_pred_valid)
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
model = RandomForestRegressor(
    n_estimators=360,
    max_depth=32,
    max_features=0.6,
    min_samples_leaf=2,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)
model.fit(X_scaled, y)
scores = cross_val_score(model, X_scaled, y, scoring='r2', cv=10, n_jobs=-1)
```

```python
np.mean(scores), np.std(scores)
```
